```python
import numpy as np
import pandas as pd
from efficient_apriori import apriori

# ladataan aineisto
df = pd.read_csv("Groceries_cleaned.csv")
values = df['itemDescription'].value_counts()

transactions = []
# ryhmitellään tuotteet ostotapahtumien kanssa
for i in range(1, 14963):
    data = df[df['InvoiceNum'] == i]['itemDescription']
    transactions.append(data)

print(transactions)
# Tehdään apriori malli
itemsets, rules = apriori(transactions, min_support=0.005, min_confidence=0)

print(len(rules))
for rule in rules:
    print(rule)
```
